###Задание 2: Подсчитайте топ10 самых популярных лем в рамках data

```python
list_data = []
for i in data_tok[:100]:
    list_data += i

dict_data = {}
for i in list_data:
    lemm = lemmatizer.lemmatize(i)
    if lemm not in dict_data:
        dict_data[lemm] = dict_data.get(lemm, 1)
    dict_data[lemm] += 1

top = []
for i in dict_data:
    top += [[i, dict_data[i]]]

sorted(top, key=lambda x: x[1], reverse = True)[:10]
```

###Задание 3: Подсчитайте количество разных слов до и после лемматизации

```python
# Чтение файла и преобразование его в список строк
with open("E:/git_exercise/quora.txt", encoding="utf-8") as f:
    data2 = f.read().split()

# Создание списка уникальных слов до лемматизации
unique_words_before = set(data2)

# Создание лемматизатора
lemmatizer = WordNetLemmatizer()

# Лемматизация каждого слова в списке
lemmatized_words = [lemmatizer.lemmatize(word) for word in data2]

# Создание списка уникальных слов после лемматизации
unique_words_after = set(lemmatized_words)

# Подсчет количества уникальных слов
print("Количество уникальных слов до лемматизации:", len(unique_words_before))
print("Количество уникальных слов после лемматизации:", len(unique_words_after))
```

###Задание 4: Подсчитайте количество разных слов до и после стемминга

```python
#ps.stem('cats')
stemmed_words = [ps.stem(word) for word in data2]

# Создание списка уникальных слов после стемминга
unique_words_after = set(stemmed_words)

# Подсчет количества уникальных слов
print("Количество уникальных слов до стемминга:", len(unique_words_before))
print("Количество уникальных слов после стемминга:", len(unique_words_after))
```

```python
def vectorize(tokens):
    ''' This function takes list of words in a sentence as input
    and returns a vector of size of filtered_vocab.It puts 0 if the
    word is not present in tokens and count of token if present.'''
    vector=[]
    for w in filtered_vocab:
        vector.append(tokens.count(w))
    return vector
def unique(sequence):
    '''This functions returns a list in which the order remains
    same and no item repeats.Using the set() function does not
    preserve the original ordering,so i didnt use that instead'''
    seen = set()
    return [x for x in sequence if not (x in seen or seen.add(x))]

#create a list of stopwords.You can import stopwords from nltk too
stopwords=["to","is","a"]

#list of special characters.You can use regular expressions too
special_char=[",",":"," ",";",".","?"]

#Write the sentences in the corpus,in our case, just two
string1="Welcome to Great Learning , Now start learning"
string2="Learning is a good practice"

#convert them to lower case
string1=string1.lower()
string2=string2.lower()

#split the sentences into tokens
tokens1=string1.split()
tokens2=string2.split()
print(tokens1)
print(tokens2)

#create a vocabulary list
vocab=unique(tokens1+tokens2)
print(vocab)

#filter the vocabulary list
filtered_vocab=[]
for w in vocab:
    if w not in stopwords and w not in special_char:
        filtered_vocab.append(w)
print(filtered_vocab)

#convert sentences into vectords
vector1=vectorize(tokens1)
print(vector1)
vector2=vectorize(tokens2)
print(vector2)
```
