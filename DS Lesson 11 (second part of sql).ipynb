Напишите влияет ли кол-во комнат на стоимость и как сильно.

```python
tasks+=1
pd.read_sql(
    '''

with cheap25 as (SELECT avg(RM) avgc25 FROM (SELECT MEDV,RM FROM boston order by MEDV LIMIT 25)),
    price_cheap25 as (SELECT avg(MEDV) pavg25 FROM (SELECT MEDV,RM FROM boston order by MEDV LIMIT 25)),
    expensive15 as (select avg(RM) avgRM15 from boston where medv=(select max(medv) from boston)),
    price_expensive15 as (select avg(MEDV) pavgEXP15 from boston where medv=(select max(medv) from boston)),    
   difference25 as (SELECT * FROM expensive15 UNION ALL SELECT *  FROM cheap25),
   diff_25 as (select (MAX(avgRM15)-MIN(avgRM15)) diff25 FROM difference25),
  
   cheap50 as (SELECT avg(RM) avgc50 FROM (SELECT MEDV,RM FROM boston order by MEDV LIMIT 50)),
   price_cheap50 as (SELECT avg(MEDV) pavg50 FROM (SELECT MEDV,RM FROM boston order by MEDV LIMIT 50)),
   difference50 as (SELECT * FROM expensive15 UNION ALL SELECT * FROM cheap50),
   diff_50 as (select (MAX(avgRM15)-MIN(avgRM15)) diff50 FROM difference50),

  

   cheap100 as (SELECT avg(RM) avgc100 FROM (SELECT MEDV,RM FROM boston order by MEDV LIMIT 100)),
    price_cheap100 as (SELECT avg(MEDV) pavg100 FROM (SELECT MEDV,RM FROM boston order by MEDV LIMIT 100)),
   difference100 as (SELECT * FROM expensive15 UNION ALL SELECT *  FROM cheap100),
   diff_100 as (select (MAX(avgRM15)-MIN(avgRM15)) diff100 FROM difference100),

  

   cheap200 as (SELECT avg(RM) avgc200 FROM (SELECT MEDV,RM FROM boston order by MEDV LIMIT 200)),
    price_cheap200 as (SELECT avg(MEDV) pavg200 FROM (SELECT MEDV,RM FROM boston order by MEDV LIMIT 200)),
   difference200 as (SELECT * FROM expensive15 UNION ALL SELECT *  FROM cheap200),
   diff_200 as (select (MAX(avgRM15)-MIN(avgRM15)) diff200 FROM difference200),

  

   cheap300 as (SELECT avg(RM) avgc300 FROM (SELECT MEDV,RM FROM boston order by MEDV LIMIT 300)),
    price_cheap300 as (SELECT avg(MEDV) pavg300 FROM (SELECT MEDV,RM FROM boston order by MEDV LIMIT 300)),
   difference300 as (SELECT * FROM expensive15 UNION ALL SELECT *  FROM cheap300),
   diff_300 as (select (MAX(avgRM15)-MIN(avgRM15)) diff300 FROM difference300),

  
  table1 as (SELECT 'Houses','Average_rooms','Average_price','Rooms_Difference' union all
  SELECT 'Expensive15',(select avgRM15 from expensive15), (select pavgEXP15 from price_expensive15), '0' union all
  SELECT 'Cheap25',(select avgc25 from cheap25), (select pavg25 from price_cheap25), (select diff25 from diff_25) union all
  SELECT 'Cheap50',(select avgc50 from cheap50), (select pavg50 from price_cheap50), (select diff50 from diff_50) union all
  SELECT 'Cheap100',(select avgc100 from cheap100), (select pavg100 from price_cheap100), (select diff100 from diff_100) union all
  SELECT 'Cheap200',(select avgc200 from cheap200), (select pavg200 from price_cheap200), (select diff200 from diff_200) union all
  SELECT 'Cheap300',(select avgc300 from cheap300), (select pavg300 from price_cheap300), (select diff300 from diff_300))
  

  Select * from table1

    ''',
    con,
)
```

7. Выведите все колонки, у которых среднее значение выше, когда дом граничит с рекой (название колонки) выбирая из CRIM, ZN, INDUS, CHAS, NOX. Напишите какой вывод можно сделать по этим данным.

```python
tasks+=1
pd.read_sql(
    ''' 
    with ycrim as (SELECT avg(CRIM) yescrim FROM boston where chas=1),
    ncrim as (SELECT avg(CRIM) nocrim FROM boston where chas=0),
    yzn as (SELECT avg(ZN) yeszn FROM boston where chas=1),
    nzn as (SELECT avg(ZN) nozn FROM boston where chas=0),
    yindus as (SELECT avg(INDUS) yesindus FROM boston where chas=1),
    nindus as (SELECT avg(INDUS) noindus FROM boston where chas=0),
    ychas as (SELECT avg(CHAS) yeschas FROM boston where chas=1),
    nchas as (SELECT avg(CHAS) nochas FROM boston where chas=0),
    ynox as (SELECT avg(NOX) yesnox FROM boston where chas=1),
    nnox as (SELECT avg(NOX) nonox FROM boston where chas=0),
  
    diff_crim as (SELECT (ycrim.yescrim-ncrim.nocrim) dc from ycrim join ncrim),
    diff_zn as (SELECT (yzn.yeszn-nzn.nozn) dz from yzn join nzn),
    diff_indus as (SELECT (yindus.yesindus-nindus.noindus) di from yindus join nindus),
    diff_chas as (SELECT (ychas.yeschas-nchas.nochas) dch from ychas join nchas),
    diff_nox as (SELECT (ynox.yesnox-nnox.nonox) dn from ynox join nnox),
    
    crim1 as (SELECT 'CRIM' colum, IIF(sign(dc)>0, 'yes','no') condition from diff_crim),
    zn1 as (SELECT 'ZN' colum, IIF(sign(dz)>0, 'yes','no') condition from diff_zn), 
    indus1 as (SELECT 'INDUS' colum, IIF(sign(di)>0, 'yes','no') condition from diff_indus),
    chas1 as (SELECT 'CHAS' colum, IIF(sign(dch)>0, 'yes','no') condition from diff_chas),
    nox1 as (SELECT 'NOX' colum, IIF(sign(dn)>0, 'yes','no') condition from diff_nox)
    
    SELECT colum FROM (
    SELECT * from crim1 
    UNION ALL 
    SELECT * from zn1
    UNION ALL 
    SELECT * from indus1
    UNION ALL 
    SELECT * from chas1
    UNION ALL 
    SELECT * from nox1
    ) WHERE condition='yes'

    ''',
    con,
)
```

```python
В домах, расположенных у реки, выше средние показатели концентрации оксидов азота и доли акров неторгового бизнеса, 
по сравнению с домами, не граничащими с рекой. По наличию оксидов азота можно предположить большее количество личных 
автомобилей у жильцов домов у реки, и более высокую опасность для проживания
```

8. Выведите значения долей промышленной застройки(INDUS), концентрации оксидов азота(NOX) и по их перцентилям - 10, 20 ... 100 ( перцетиль(10,20...100),значение INDUS, значение NOX). Напишите прослеживается между ними взаимосвязь

```python
tasks+=1
pd.read_sql(
    '''
WITH p_indus AS (SELECT INDUS, NTILE(10) OVER (ORDER BY INDUS) AS percentile1 FROM boston),
p_nox AS (SELECT NOX, NTILE(10) OVER (ORDER BY NOX) AS percentile2 FROM boston),
table1 as (SELECT percentile1, MAX(INDUS) as INDUS1 FROM p_indus GROUP BY percentile1),
table2 as (SELECT percentile2, MAX(NOX) as NOX1 FROM p_nox GROUP BY percentile2)
SELECT t1.percentile1 percentile, t1.INDUS1,t2.NOX1 FROM table1 t1 JOIN table2 t2 on t1.percentile1=t2.percentile2

    ''',
    con,
)
```
