8. Какая разница в стоимости между самым дорогим и самым дешевым заказом из США.

```python
tasks+=1
pd.read_sql(
    '''
    /*Объединяем Customers и Order в CustomersOrders*/
    with qCusOr as (SELECT c.CustomerID, o.OrderID FROM Customers c
    join Orders o on c.CustomerID=o.CustomerID where c.Country='USA'),
    CusOr as (SELECT * FROM qCusOr),
    /*Объединяем OrderDetails и CustomersOrders*/
    qCusOrOd as (SELECT od.ProductID,od.Quantity, co.CustomerID FROM OrderDetails od 
    join CusOR co on od.OrderId=co.OrderID),
    CusOrOd as (SELECT * from qCusOrOd),
    /*Объединяем Products и CustomersOrdersDeatails*/
    qCusOrOdP as (SELECT p.Price,cod.Quantity FROM Products p 
    join CusOrOD cod on p.ProductID=cod.ProductID),
    CusOrOdP as (SELECT * FROM qCusOrOdP)
    /*Считаем разницу*/
    SELECT (MAX(Price*Quantity)-MIN(Price*Quantity)) Difference FROM CusOrOdP
    ''',
    con,
)
```

9. Выведите количество заказов у каждого их трех самых молодых сотрудников, а также имя и фамилию во второй колонке.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT count(o.OrderID) number_of_orders, em.FirstName|| ' '||em.LastName CommonName, em.BirthDate FROM Orders o 
    join (SELECT EmployeeID,FirstName,LastName, BirthDate,
    (substr(BirthDate,7,4)||'-'||substr(BirthDate,4,2)||'-'||substr(BirthDate,1,2)) as new_date
    FROM Employees 
    ORDER BY new_date DESC
    LIMIT 3) em on o.EmployeeID=em.EmployeeID
    GROUP BY o.EmployeeID

    ''',
    con,
)
```
