```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import cauchy, norm
import math
import itertools
%matplotlib inline
```

```python
cauchy_rv = sts.cauchy()
sample = cauchy_rv.rvs(50)
```

```python
plt.hist(sample, density=True)
x = np.linspace(-10,10,100)
pdf = cauchy_rv.pdf(x)
plt.plot(x, pdf, label='теоретическая плотность', alpha=0.9)
plt.legend()
```

```python
#plt.figure(figsize=(16,9))
# сгенерируем выборочные средние
'''
for N in [5, 10, 20, 25, 30]:
    means = []
    for i in range(100):
         means.append(np.mean(cauchy_rv.rvs(N)))
    plt.hist(means, alpha=0.5)

    for M in means:
        norm_rv = sts.norm(loc=M.mean(), scale=M.std() / np.sqrt(N))
        x = np.linspace(1,200, 100)
        pdf = norm_rv.pdf(x)
    plt.hist(x, pdf, label='Теоретическая плотность', alpha=0.5)
plt.legend(['5', '10', '20', '25', '30'])
plt.show()
'''
df=50
size = [5,10,50]
sm = []
bins = np.arange(5.,100.,0.5)
for i in size:
    for _ in itertools.repeat(None, 1000):
        rv = cauchy.rvs(df, size=i)
        sm.append(np.mean(rv))
# построим гистограмму и график нормального распределения для выборки размера 5
plt.hist(sm[0:200], 20, density=True, histtype='stepfilled', alpha=0.7, label='mean hist')
plt.plot(bins, norm.pdf(bins, df, np.sqrt(2*df/5.)),'r', alpha=0.9, label='norm pdf')
plt.xlabel('Sample mean')
plt.ylabel('Probability')
```
