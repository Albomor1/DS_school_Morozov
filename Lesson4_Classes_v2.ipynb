```python
#Вариант 1
import pickle
import matplotlib.pyplot as plt
import pandas as pd

class PickleReader:
    def __init__(self, file_path):
        self.file_path = file_path

    def read_pickle(self):
        with open(self.file_path, 'rb') as pickle_file:
            data = pickle.load(pickle_file)
            return data

class PickleGraph:
    def __init__(self, df):
        self.df = df

    def plot_graph(self):
        #plt.plot(self.df['region'], self.df['objSquareLiving'])
        grouped = self.df.groupby('region')
        for name, group in grouped:
            plt.plot(group['region'], group['objSquareLiving'], label=name)
        plt.xlabel('region')
        plt.ylabel('objSquareLiving')
        plt.show()

# Создание экземпляра класса
pickle_reader = PickleReader('E:/git_exercise/df_nashdomrf.pkl')
df = pd.DataFrame(pickle_reader.read_pickle())
pickle_graph = PickleGraph(df)

# Построение графика
pickle_graph.plot_graph()
```

```python
#Черновик
import csv
import matplotlib.pyplot as plt

class CSVReader:
    def __init__(self, file_path):
        self.file_path = file_path

    def read_csv(self):
        with open(self.file_path, 'r') as csv_file:
            reader = csv.reader(csv_file, delimiter=';')
            x_values = []
            y_values = []
            for row in reader:
                x_values.append(row[0])
                y_values.append(row[1])
            return x_values, y_values

class CSVGraph:
    def __init__(self, x_values, y_values, title):
        self.x_values = x_values
        self.y_values = y_values
        self.title = title

    def plot_graph(self):
        plt.plot(self.x_values, self.y_values)
        plt.xlabel('X Axis')
        plt.ylabel('Y Axis')
        plt.title(self.title)
        plt.show()

# Создание экземпляра класса
csv_reader = CSVReader('E:/git_exercise/class_df.csv')
csv_graph = CSVGraph(csv_reader.read_csv()[0], csv_reader.read_csv()[1], 'Title of the Graph')

# Построение графика
csv_graph.plot_graph()
```
